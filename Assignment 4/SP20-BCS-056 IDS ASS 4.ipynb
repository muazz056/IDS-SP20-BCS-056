```python
import pandas as pd
import numpy as np
```

```python
# Read the dataset
df1 = pd.read_csv("G-prediction.csv")
df1
```

```python
# Import LabelEncoder
from sklearn import preprocessing

# Creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
df1['beard']=le.fit_transform(df1['beard'])
df1['hair_length']=le.fit_transform(df1['hair_length'])
df1['scarf']=le.fit_transform(df1['scarf'])
df1['eye_color']=le.fit_transform(df1['eye_color'])
df1['gender']=le.fit_transform(df1['gender'])
df1
```

```python
x = df1[['height','weight','beard','hair_length','shoe_size','scarf','eye_color']]
y = df1['gender']

#Spliting Data for traning(67%) and testing(33%)
from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=42)

#Spliting Data for traning(80%) and testing(20%)
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
# Import MLPClassifer 
from sklearn.neural_network import MLPClassifier

# Create model object
clf = MLPClassifier(hidden_layer_sizes=(6,5),
                    random_state=5,
                    verbose=True,
                    learning_rate_init=0.01)

# Fit data onto the model
clf.fit(X_train,y_train)
```

```python
# Make prediction on test dataset
ypred=clf.predict(X_test)

# Import accuracy score 
from sklearn.metrics import accuracy_score

# Calcuate accuracy
accuracy_score(y_test,ypred)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,ypred))

print(classification_report(y_test,ypred))
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
rf_clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
rf_clf.fit(X_train,y_train)

random_y_pred=rf_clf.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,random_y_pred))

print(classification_report(y_test,random_y_pred))
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
svm_clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
svm_clf.fit(X_train, y_train)

#Predict the response for test dataset
svm_pred = svm_clf.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,svm_pred))

print(classification_report(y_test,svm_pred))
```

```python

#With MonteCarlo cross validation

from sklearn.model_selection import ShuffleSplit,cross_val_score
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
shuffle_split=ShuffleSplit(test_size=0.3,train_size=0.5,n_splits=10)

scores=cross_val_score(dt,x,y,scoring="f1",cv=shuffle_split)

print("cross Validation scores:n {}".format(scores))
print("Average Cross Validation score :{}".format(scores.mean()))

```

```python
#With P-Out cross validation
from sklearn.model_selection import LeavePOut,cross_val_score
from sklearn.tree import DecisionTreeClassifier

#  Ignore the unnecessary warnings
import warnings
warnings.filterwarnings('ignore')

lpo=LeavePOut(p=2)
lpo.get_n_splits(x)
tree=DecisionTreeClassifier(max_depth=5)
score=cross_val_score(tree,x,y,scoring="f1",cv=lpo)
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```
